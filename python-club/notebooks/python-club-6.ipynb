##Exercise 6.0
When could we get into an infinite loop with the fibonnaci code below? What data type do we require for the parameters? Add some additional conditions (guardians) to the code to prevent such issues.

```
def fibonacci (n):
    if n == 0:
        return 0
    elif  n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```
def first(word):
    return word[0]

def last(word):
    return word[-1]

def middle(word):
    return word[1:-1]
```
